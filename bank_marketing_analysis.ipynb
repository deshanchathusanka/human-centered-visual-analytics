# &nbsp;$\color{Blue}{\text{3. Implementation}}$

```python
selected_features = ['poutcome_cat', 'age', 'default_cat', 'duration', 'previous','pdays','cons_conf_idx']
selected_df = df[selected_features]
selected_df
df_norm = (selected_df - selected_df.min())/(selected_df.max() - selected_df.min())
df_norm['y'] = df['y']

parallel = alt.Chart(data = df_norm,
          title = 'Parallel Coordinate Visualization').transform_window(
    index='count()'
).transform_fold(
    selected_features
).mark_line().encode(
    alt.X('key:N',sort = None),
    y='value:Q',
    color = alt.Color('y:N', scale = alt.Scale(domain = ['yes', 'no'],
                      range = ['red', 'green'])),
    detail='index:N',
    opacity=alt.value(0.5)
).properties(
    width = 1200
)
parallel
```

```python
poutcome_df = df[['poutcome', 'poutcome_cat']].drop_duplicates()
poutcome_df['poutcome_cat_norm'] = (poutcome_df['poutcome_cat']-poutcome_df['poutcome_cat'].min())\
                                    /(poutcome_df['poutcome_cat'].max() - poutcome_df['poutcome_cat'].min())
poutcome_df.sort_values(by = 'poutcome_cat', inplace = True)
poutcome_df
```

```python
default_df = df[['default','default_cat']].drop_duplicates()
default_df['default_cat_norm'] = (default_df['default_cat']-default_df['default_cat'].min())\
/(default_df['default_cat'].max()-default_df['default_cat'].min())
default_df.sort_values(by = 'default_cat', inplace = True)
default_df
```

${\color{Red}{\text{Prominent features :}}
\color{Blue}{\text{Age, Consumer confident index, Duration, Previous outcome, Default category, Previous, Pdays}
}}$

```python
import statistics

stat_df = pd.DataFrame(columns = ['feature', 'y', 'mode(cat)/mean(num)'])
cat_features = ['poutcome_cat', 'default_cat']

for index in selected_features:

  if index in cat_features:
    empty_row = {
        'feature' : index,
        'y' : 'yes',
        'mode(cat)/mean(num)' : 0
    }
    stat_df = stat_df.append(empty_row, ignore_index=True)

    empty_row = {
        'feature' : index,
        'y' : 'no',
        'mode(cat)/mean(num)' : 0
    }
    stat_df = stat_df.append(empty_row, ignore_index=True)
  else:
    feature_y_no = df_norm[df_norm['y']=='no']
    feature_y_no_mode = statistics.mean(feature_y_no[index])
    feature_y_no_row = {
        'feature' : index,
        'y' : 'no',
        'mode(cat)/mean(num)' : feature_y_no_mode
    }
    stat_df = stat_df.append(feature_y_no_row, ignore_index=True)

    feature_y_yes = df_norm[df_norm['y']=='yes']
    feature_y_yes_mode = statistics.mean(feature_y_yes[index])
    feature_y_yes_row = {
        'feature' : index,
        'y' : 'yes',
        'mode(cat)/mean(num)' : feature_y_yes_mode
    }
    stat_df = stat_df.append(feature_y_yes_row, ignore_index=True)

stat_df
```

```python
from tables.table import Column
bar = alt.Chart(stat_df, title = 'Bar Graph Representation').mark_bar().encode(
    alt.X('y:N'),
    alt.Y('mode(cat)/mean(num):Q', title = 'Average'),
    alt.Color('y:N'),
    alt.Column('feature:N', sort = selected_features)
).properties(
    width = 150
)

alt.vconcat(parallel, bar)
```

```python
color = alt.Color('y:N', scale = alt.Scale(domain = ['yes', 'no'],
                      range = ['red', 'green']))
click = alt.selection_multi(encodings=['color'])

parallel = alt.Chart(data = df_norm, title = 'Parallel Coordinate Representation').transform_window(
    index='count()'
).transform_fold(
    selected_features
).mark_line().encode(
    alt.X('key:N',sort = None),
    y='value:Q',
    color = alt.condition(click, color, alt.value('lightgray')),
    detail='index:N',
    opacity=alt.value(0.5)
).add_selection(
    click
).properties(
    width = 1200
)

bar = alt.Chart(stat_df, title = 'Numerical Field Representation').mark_bar().encode(
    alt.X('y:N'),
    alt.Y('mode(cat)/mean(num):Q', title = 'Average'),
    alt.Column('feature:N', sort = selected_features),
    color = alt.condition(click, color, alt.value('lightgray')),
).add_selection(
    click
).properties(
    width = 150
)

alt.vconcat(parallel, bar)
```

```python
poutcome_cat_bar = alt.Chart(df_norm, title = 'Categorical Field Representation').mark_bar().encode(
    x = 'count():Q',
    y = 'y:N',
    color = alt.condition(click, color, alt.value('lightgray')),
    row = alt.Row('poutcome_cat:N', sort='descending')
).add_selection(
    click
).properties(
    height = 85,
    width = 120
)

default_cat_bar = alt.Chart(df_norm, title = 'Categorical Field Representation').mark_bar().encode(
    x = 'count():Q',
    y = 'y:N',
    color = alt.condition(click, color, alt.value('lightgray')),
    row = alt.Row('default_cat:N', sort='descending')
).add_selection(
    click
).properties(
    height = 85,
    width = 120
)

alt.vconcat(parallel, bar) | poutcome_cat_bar
```
