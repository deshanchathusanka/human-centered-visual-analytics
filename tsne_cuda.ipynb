```python
from tsnecuda import TSNE
import pandas as pd
import numpy as np
import altair as alt
import time
```

```python
mnist_df = pd.read_csv('/content/data/mnist/mnist_test.csv')
x_df = mnist_df.drop(['label'], axis = 1)
y_df = mnist_df['label']
```

```python
mnist_embedded = TSNE(n_components = 2, perplexity = 15, learning_rate = 10).fit_transform(x_df)
mnist_embedded_df = pd.DataFrame({
    'tsne_1':mnist_embedded[:, 0],
    'tsne_2':mnist_embedded[:, 1],
    'y':y_df
})

alt.data_transformers.disable_max_rows()
scatter = alt.Chart(mnist_embedded_df).mark_circle().encode(
    x = 'tsne_1:Q',
    y = 'tsne_2:Q',
    color = 'y:N'
).interactive()
scatter
```

```python
tsne_cuda_performance = []
for k in range(0, 100):
  start = time.perf_counter()
  mnist_embedded = TSNE(n_components = 2, perplexity = 15, learning_rate = 10).fit_transform(x_df)
  end = time.perf_counter()

  execution_time = end - start
  tsne_cuda_performance.append(execution_time)

tsne_cuda_performance_df = pd.DataFrame({
    'method' : ['tsne_cuda' for i in range(100)],
    'exe_time': tsne_cuda_performance
})
box_plot = alt.Chart(tsne_cuda_performance_df).mark_boxplot(extent='min-max').encode(
    x = 'method:N',
    y = alt.Y('exe_time:Q', scale=alt.Scale(domain=(3.40, 3.85)))
)
box_plot
```
